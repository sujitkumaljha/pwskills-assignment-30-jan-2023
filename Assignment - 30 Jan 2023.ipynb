While loop is used when we have to stop/start executing some codes only when some condition(s) is/are fultilled. e.g 'Count from 1 to 10' - Here the conditionn is to start from int 1 and keep executing untill the counter doesn't reach 10 or in other words keep executing from 1 to till the counter is less than 11.

```python
# 1st pattern
i = int(5)
while (i > 0):
    j = int(0)
    while (j < i):
        print(j+1,end='')
        j+=1
    i -= 1
    print("")
```

```python
# 2nd Pattern
i = int(5)
count = int(1)
while (i > 0):
    j = i - count
    while (j < i):
        print(i - j, end='')
        j += 1
    i -= 1
    count += 1
    print("")
```

```python
# 3rd Pattern
i = int(5)
while (i > 0):
    j = i
    while (j > 0):
        print(j, end='')
        j -= 1
    print("")
    i -= 1
```
